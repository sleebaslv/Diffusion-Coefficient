```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import *

data2 = np.loadtxt('diff-coeff-field-final.dat')
field = data2[:,0]
x_fld = data2[:,1]
y_fld = data2[:,2]
z_fld = data2[:,3]

fig, axs = plt.subplots(1,3,figsize=(18,5))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=None)

axs[0].plot(field, x_fld, '-ob')
axs[0].set_xlabel('Field')
axs[0].set_ylabel('Diffusion Coefficient')
axs[0].set_title('X-Direction')
#axs[0].set_yscale('log', basey=10)
axs[0].set_xscale('log', basex=10)


axs[1].plot(field, y_fld, '-ob')
axs[1].set_xlabel('Field')
axs[1].set_ylabel('Diffusion Coefficient')
axs[1].set_title('Y-Direction')
#axs[1].set_yscale('log', basey=10)
axs[1].set_xscale('log', basex=10)


axs[2].plot(field, z_fld, '-ob')
axs[2].set_xlabel('Field')
axs[2].set_ylabel('Diffusion Coefficient')
axs[2].set_title('Z-Direction')
#axs[2].set_yscale('log', basey=10)
axs[2].set_xscale('log', basex=10)
plt.savefig('dc.pdf')
plt.show()
```
